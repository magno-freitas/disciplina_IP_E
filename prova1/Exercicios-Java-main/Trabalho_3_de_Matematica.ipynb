```python
import matplotlib.pyplot as plt
from matplotlib_venn import venn3

print("Nome: Leonardo Cesar Caixeta Romagnoli")

# CONJUNTOS

hardware = {"Microsoft", "Apple", "IBM", "Tesla", "NVIDIA"}
cloud = {"Google", "Microsoft", "Apple", "Amazon", "Meta", "IBM", "NVIDIA", "Oracle"}
redes = {"Google", "Meta", "TikTok"}

# Adicionando novos elementos
cloud.add("Alibaba")
redes.add("Twitter")

# OPERAÇÕES

# União
uniao = hardware.union(cloud).union(redes)

# Pergunta 1:
# Quais empresas atuam em pelo menos um segmento?
print("\nResposta 1:")
print("Empresas:", uniao)
print("Total:", len(uniao))

# Interseção
intersecao = hardware.intersection(cloud)

# Pergunta 2:
# Quais empresas atuam em hardware e cloud ao mesmo tempo?
print("\nResposta 2:")
print("Empresas:", intersecao)
print("Total:", len(intersecao))

# Diferença
diferenca = hardware.difference(cloud).difference(redes)

# Pergunta 3:
# Quais empresas atuam apenas em hardware?
print("\nResposta 3:")
print("Empresas:", diferenca)
print("Total:", len(diferenca))

# DIAGRAMA DE VENN

plt.figure()
venn3([hardware, cloud, redes], ('Hardware', 'Cloud', 'Redes'))
plt.title("Diagrama de Venn - Empresas")
plt.show()
```

```python
# PARTE 2.1

matematica = {"João","Lia","Marcos","Nina","Otávio","Paula","Rafa","Amanda","Bianca","Caio","Daniel","Eduarda","Fernanda","Gustavo","Heloísa"}

fisica = {"Lia","Marcos","Sofia","Otávio","Paula","Tiago","Bianca","Caio","Eduardo","Fernanda","Gabriel","Heloísa","Isabela","Juliano","Karen"}

# a) Interseção
ambos = matematica.intersection(fisica)
print("\nResposta A:")
print("Estudantes:", ambos)
print("Total:", len(ambos))

# b) Só matemática
so_matematica = matematica.difference(fisica)
print("\nResposta B:")
print("Estudantes:", so_matematica)
print("Total:", len(so_matematica))

# c) Só física
so_fisica = fisica.difference(matematica)
print("\nResposta C:")
print("Estudantes:", so_fisica)
print("Total:", len(so_fisica))

# d) União
pelo_menos_uma = matematica.union(fisica)
print("\nResposta D:")
print("Estudantes:", pelo_menos_uma)
print("Total:", len(pelo_menos_uma))

# TROCA (EXERCÍCIO)

print("\n--- TROCA (EXERCÍCIO) ---")
print("Usando união no lugar da interseção:", matematica.union(fisica))
print("Usando interseção no lugar da união:", matematica.intersection(fisica))

# PARTE 2.2

A = {0, 1, 2}             # (-1, 2]
B = {-1, 0, 1, 2, 3, 4}   # (-2, 4]
C = {0, 1, 2}             # [0, 3)

# (A ∩ B)
intersecao_AB = A.intersection(B)

# (A ∩ B) ∪ C
uniao = intersecao_AB.union(C)

# (B ∩ C)
intersecao_BC = B.intersection(C)

# resultado final
resultado = uniao.difference(intersecao_BC)

print("\n--- PARTE 2.2 ---")
print("\nResultado: ", resultado)
print("Resultado matematicamente certo: (-1, 0)")
```
